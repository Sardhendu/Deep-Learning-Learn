```python
# Class encapculating the code for plot
class plot():
    def __init__(self, rows, columns, flag_size = (15,5)):
        self.fig, self.axs = plt.subplots(rows,columns, figsize=flag_size, facecolor='w', edgecolor='k')
        self.axs = self.axs.ravel()
        
    def plot_images(self, plot_no, img, label=None): 
        self.axs[plot_no].imshow(img)
        if label:
            self.axs[plot_no].set_title(label)

    def disp_plots(self, img_dataset, img_labels):  # Give the function the complete dataset (or the sampled)
        for i, image in enumerate(img_dataset):
            self.plot_images(i, image, img_labels[i])
```

```python
# Before we build the Model lets first retrieve out data set that we created and stored in the "data preparation" step.
fnl_dataset_path = '/Users/sam/All-Program/App-DataSet/Deep-Neural-Nets/MNIST_ImageClassification/Datapreparation/dataset_complete.p'
with open(fnl_dataset_path, 'rb') as f:
    fnl_dataset = pickle.load(f)
    training_dataset = (fnl_dataset['training_dataset'])
    training_labels = (fnl_dataset['training_labels'])
    test_dataset = (fnl_dataset['test_dataset'])
    test_labels = (fnl_dataset['test_labels'])
    crossvalid_dataset = (fnl_dataset['crossvalid_dataset'])
    crossvalid_labels = (fnl_dataset['crossvalid_labels'])
    
```

```python
# Check redundancy in the dataset, since the data is not a manually clensed, we check if the same data(image)
# is present in both the training and test set. If so then, our model may be succeptible to overfitting.

# Using numpy arrays and comparing each dataset
def capture_overlaps_px(dataset1, dataset2):
    overlaping_imgs = defaultdict(list)
    for i , img1 in enumerate(dataset1):
        for j, img2 in enumerate(dataset2):
            if np.array_equal(img1, img2):
                overlaping_imgs[i].append(j)
    return overlaping_imgs

# Now we would like to get rid of the overlaping images, for the entire test set  
def capture_overlaps_hv(dataset1, dataset2, labels = None, delete=None):
    hashval_1 = np.array([hashlib.sha256(img).hexdigest() for img in dataset1])
    hashval_2 = np.array([hashlib.sha256(img).hexdigest() for img in dataset2])
    overlaping_imgs = {} # list of indexes
    for i, hash1 in enumerate(hashval_1):
        overlapping_indices = np.where(hashval_2 == hash1)
        if len(overlapping_indices[0]):
            overlaping_imgs[i] = list(overlapping_indices[0])#.append([i]+list(overlapping_indices[0])) 
    if delete:
        overlap = list(overlaping_imgs.keys())
        return np.delete(dataset1, overlap, 0), np.delete(labels, overlap, None)
    else:
        return overlaping_imgs

print ('Overall time taken by using pixel arrays and numpy array for comparison ...........')
%time overlaping_imgs_px = capture_overlaps_px(test_dataset[0:50], training_dataset)
print ('Overall time taken by using hash function and numpy array for comparison ..........')
%time overlaping_imgs_hv = capture_overlaps_hv(test_dataset[0:50], training_dataset)
# 
print ('')
print ('Number of overlaps using pixel comparison the given test size is: ', len(overlaping_imgs_px.keys()))
print ('Number of overlaps using hashvalue comparison the given test size is: ', len(overlaping_imgs_hv.keys()))
print ('done')
```

```python
# We see that we have received many-many overlapping dataset between the training and test samples. Let plot few of them
print ('The overlapping images using the pixel comparison are: ', overlaping_imgs_px)
print ('The overlapping images using the hashvalue comparison are: ', overlaping_imgs_hv)

print ('')

overlapping_imgs = overlaping_imgs_hv
random_img = random.sample(list(overlapping_imgs.keys()),10)
print ('Below are few randomly selected images that Overlap:', random_img)

obj_plot = plot(rows=5, columns=8)
cnt = 0


for common_imgs in random_img:
    target = overlapping_imgs[common_imgs]
    obj_plot.plot_images(cnt, test_dataset[common_imgs])   #common_imgs[0]
    cnt += 1
    for imgs in target:
        obj_plot.plot_images(cnt, training_dataset[imgs])
        cnt += 1
    
    

    
```

```python
# We see that the hash technique is much faster than the pixel comparison technique, so we use the hash technique to 
# remove all the images that overlaps (present in both training and test set).

# Removing duplicaties from the entire test_dataset
print ('count of test dataset before :', len(test_labels))
test_dataset_fnl, test_labels_fnl = capture_overlaps_hv(test_dataset, training_dataset, test_labels, delete='yes')
print ('count of test dataset after :', len(test_labels_fnl))

# Removing duplicaties from the entire crossvalid_dataset
print ('count of crossvalid dataset before :', len(crossvalid_labels))
crossvalid_dataset_fnl, crossvalid_labels_fnl = capture_overlaps_hv(crossvalid_dataset, training_dataset, crossvalid_labels, delete='yes')
print ('count of crossvalid dataset after :', len(crossvalid_labels_fnl))

# storing the new test_data set and crossvalid_dataset into the disk
# Now we simply store the training, crossvalid and test dataset into the disk

cleaned_dataset_path = '/Users/sam/All-Program/App-DataSet/Deep-Neural-Nets/MNIST_ImageClassification/Datapreparation/dataset_cleaned.p'

try:
    f = open(cleaned_dataset_path, 'wb')
    dataset_cleaned = {
        'training_dataset': training_dataset,
        'training_labels': training_labels,
        'crossvalid_dataset': crossvalid_dataset_fnl,
        'crossvalid_labels': crossvalid_labels_fnl,
        'test_dataset': test_dataset_fnl,
        'test_labels': test_labels_fnl,
    }
    pickle.dump(dataset_cleaned, f, pickle.HIGHEST_PROTOCOL)
    f.close()
except Exception as e:
    print('Unable to save data to', complete_dataset_path, ':', e)
    raise

    
statinfo = os.stat(cleaned_dataset_path)
print('Compressed pickle size:', statinfo.st_size)
```
