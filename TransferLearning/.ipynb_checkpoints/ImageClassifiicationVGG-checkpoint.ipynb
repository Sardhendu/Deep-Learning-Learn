```python
%matplotlib inline

import numpy as np
import urllib
import tensorflow as tf
import scipy as spy
from matplotlib import pyplot as plt

from datasets import imagenet
from nets import vgg
from datasets import dataset_utils
from preprocessing import vgg_preprocessing

from tensorflow.python.framework import ops

slim = tf.contrib.slim
imageSize = vgg.vgg_16.default_image_size
print (imageSize)
```

### Let us now perform VGG net on one IMAGE:

```python
# Plot image and activations:

def plot(dataIN, title):
    plt.figure()
    plt.imshow(dataIN.astype(np.uint8))  # Convering into insigned integer
    plt.suptitle(title, fontsize=14, color = 'r')
    plt.axis('off')
    plt.show()
```

```python
# The image should be of same size as that of VGG trained (224x224)
imagePath = "/Users/sam/All-Program/App-DataSet/Deep-Neural-Nets/TransferLearning/dogImage.jpg"

debugg = True
ops.reset_default_graph()
sess = tf.Session()

with sess.as_default():
#     image_string = urllib.urlopen(imagePath).read()
    
#     image = tf.image.decode_jpeg(image_string, channels=3)
    # load the image usign sckitlearn
    imagePixel = spy.ndimage.imread(imagePath).astype(float)
    print (imagePixel.shape)
    plot(imagePixel, title='DOG')
    # Preprocessing the image using VGG net
    preprocessedImage = vgg_preprocessing.preprocess_image(imagePixel, imageSize, imageSize, is_training=False)
    
    
    # We have our data preprocessed, we would now have to calssify it
    batchData = tf.expand_dims(preprocessedImage, 0)
    
    # We would use the VGG default argument scope for the modeling part. The argument are the configuration
    # of the network, we could even use our custom argscope, if we would like
    with slim.arg_scope(vgg.vgg_arg_scope()):
        logits, _ = vgg.vgg_16(batchData, num_classes = 1000, is_training = False)
        
    # Getting the probability Score:
    score = tf.nn.softmax(logits)
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        ppsImage, scr = sess.run([preprocessedImage, score])
        plot(ppsImage, title='You would see that the image is a little bit cropped from the center.')
#         print(scr)
        
    sorted_inds = [i[0] for i in sorted(enumerate(-scr), key=lambda x:x[1])]
#     names = imagenet.create_readable_names_for_imagenet_labels()
#     for i in range(5):
#         index = sorted_inds[i]
#         print('Probability %0.2f => [%s]' % (scr[index], names[index+1]))
        
#     if debugg:
#         preprocessedImage = preprocessed_image.eval()
#         print (preprocessedImage.shape)
#         print(batchData.eval().shape)
#         print(logits.eval())
        
```

```python
from preprocessing.vgg_preprocessing import _mean_image_subtraction,_R_MEAN, _G_MEAN, _B_MEAN

# These are defined VGG net parameters for RGB
# _R_MEAN = 123.68
# _G_MEAN = 116.78
# _B_MEAN = 103.94
imagePath = "/Users/sam/All-Program/App-DataSet/Deep-Neural-Nets/TransferLearning/Traffic.jpg"
ops.reset_default_graph()
sess = tf.Session()

with sess.as_default():
    imagePixel = spy.ndimage.imread(imagePath).astype(float)
    imagePixel = spy.misc.imresize(imagePixel, (224, 224))
    print (imagePixel.shape)
    plot(imagePixel, title='Traffic')
    
#     Preprocessing Image:
    preprocessedImage = _mean_image_subtraction(tf.to_float(imagePixel, name="to_float"), [_R_MEAN, _G_MEAN, _B_MEAN])
    input_image = tf.expand_dims(preprocessedImage, 0)
    print ('Preprocessed Expanded Dims Shape: ', input_image.get_shape().as_list())
    
    with slim.arg_scope(vgg.vgg_arg_scope()):
        logits, _ = vgg.vgg_16(input_image,
                               num_classes=1000,
                               is_training=False,
                               spatial_squeeze=False)
    print ('Logit shape: ', logits.get_shape().as_list())
    prediction = tf.argmax(logits, dimension=3)
    
    init_fn = slim.assign_from_checkpoint_fn(os.path.join(checkpoints_dir, 'vgg_16.ckpt'),
                                             slim.get_model_variables('vgg_16'))
    print (init_fn)
    with tf.Session() as sess:
        init_fn(sess)
        pImage, lgts, pred = sess.run([preprocessedImage, logits, prediction])
#         print(pImage)
        print (pred)
        plot(pImage, title="Mean normalized - NO CROPPING OF IMAGE HERE, SINCE WE DONT USE THE VGG PREPROCESSOR")


        
predNum = tf.squeeze(pred)
print (predNum)

    
```
